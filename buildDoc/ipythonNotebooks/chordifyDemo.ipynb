```python
from music21 import *
```

```python
%load_ext music21.ipython21.ipExtension
```

```python
b = corpus.parse('bwv66.6')
```

```python
b.show()
```

```python
bChords = b.chordify()
```

```python
b.insert(0, bChords)
```

```python
b.show()
```

```python
for c in bChords.flat.getElementsByClass('Chord'):
    rn = roman.romanNumeralFromChord(c, key.Key('A'))
    c.addLyric(str(rn.figure))
```

```python
b.show()
```

```python
soprano = b.parts[0]
for n in soprano.flat.notes:
    n.addLyric(str(n.beatStrength))
```

```python
b.show()
```

```python
b.show('musicxml')
```

```python
b.plot('pianoroll')
```
